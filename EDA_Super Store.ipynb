### Branch

```python
#Import mode function:

#Impute data and check #missing values before and after imputation to confirm
print('\nOrignal #missing: %d'% data['Branch'].isnull().sum())
data['Branch'] = np.where(data['Branch'].isnull(),data['Branch'].mode(),data['Branch'])
print('Count missing after pre-processing: %d' %(data['Branch'].isnull().sum()))
```

```python
#Gender
```

```python
#Import mode function:

#Impute data and check #missing values before and after imputation to confirm
print('\nOrignal #missing: %d'% data['Gender'].isnull().sum())
data['Gender'] = np.where(data['Gender'].isnull(),data['Gender'].mode(),data['Gender'])
print('Count missing after pre-processing: %d' %(data['Gender'].isnull().sum()))
```

```python
data['Gender'].value_counts()
```

```python
#cogs
```

```python
#Import mode function:

#Impute data and check #missing values before and after imputation to confirm
print('\nOrignal #missing: %d'% data['cogs'].isnull().sum())
data['cogs'] = np.where(data['cogs'].isnull(),data['cogs'].median(),data['cogs'])
print('Count missing after pre-processing: %d' %(data['cogs'].isnull().sum()))
```

```python
data.apply(lambda x : len(x.unique()))
```

```python
#Filter categorical variables
categorical_columns = [x for x in data.dtypes.index if data.dtypes[x]=='object']
#Exclude ID cols and source:
categorical_columns = [x for x in categorical_columns if x not in ['Time','Date','Invoice ID']]          
#Print frequency of categories
for col in categorical_columns:
    print('\nFrequency of Categories for varible %s' %col)
    print(data[col].value_counts())
```

```python
Payment value counts
```

```python
#Change categories of Payment:
print('Actual Categories:')
print(data['Payment'].value_counts())

print("Updated Categories:")
data['Payment'] = data['Payment'].replace({'Csh':'Cash','wallet':'Ewallet','credit':'Credit card','credit ':'Credit card','Credit ':'Credit card','Credit Card':'Credit card'})
print(data['Payment'].value_counts())
```
